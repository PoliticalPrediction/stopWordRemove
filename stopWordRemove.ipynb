```python
import nltk
stop_words = nltk.corpus.stopwords.words('english')
set(stop_words)
```

```python
from nltk.tokenize import TweetTokenizer
tt = TweetTokenizer(    
    preserve_case=False,
    strip_handles=True,
    reduce_len=True)
tweet = "This is a cooool #dummysmiley: :-) :-P <3 and some arrows < > -> <-- @remy: This is waaaaayyyy too much for you!!!!!!"
tokenize_tweets = tt.tokenize(tweet) 

tokenize_words_without_stop_words = []
for word in tokenize_tweets:
    if word not in stop_words:
        tokenize_words_without_stop_words.append(word)
print(set(tokenize_tweets) - set(tokenize_words_without_stop_words))
print("**********************")
print(tokenize_tweets)
print("**********************")
print(tokenize_words_without_stop_words)
```

```python
import string
import re
import nltk
from nltk.tokenize import TweetTokenizer

tweet = "This is a cooool #dummysmiley: :-) :-P <3 and some arrows < > -> <-- @remy: This is waaaaayyyy too much for you!!!!!!"

def clean_text(text):
    # remove numbers
    text_nonum = re.sub(r'\d+', '', text)
    # remove punctuations and convert characters to lower case
    text_nopunct = "".join([char.lower() for char in text_nonum if char not in string.punctuation]) 
    # substitute multiple whitespace with single whitespace
    # Also, removes leading and trailing whitespaces
    text_no_doublespace = re.sub('\s+', ' ', text_nopunct).strip()
    return text_no_doublespace

cleaned_tweet = clean_text(tweet)
tt = tt.tokenize(cleaned_tweet)

tokenize_words_without_stop_words = []
for word in tt:
    if word not in stop_words:
        tokenize_words_without_stop_words.append(word)
print(set(tt) - set(tokenize_words_without_stop_words))
print("**********************")
print(tt)
print("**********************")
print(tokenize_words_without_stop_words)
```
